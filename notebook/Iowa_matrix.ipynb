```python
def read_matrix(filename):
    

    L3 = []
    for x in range(101):
        L3.append([0]*101)
    file_obj = open(filename,'r')
    for line in file_obj:
        line = line.rstrip()
        fields = line.split()
        #print test2
        #print fields
        x = int(fields[0].split('-')[0])
        y = int(fields[0].split('-')[1])
        if x<101 and y<101:
            L3[x][y] = int(fields[1])
    file_obj.close()
    #print "test"
    #print L1
    #print L2
    #print L3
    #print dimension
    
    return L3


            
            
```

```python
filename = "../data/two_files_matrix.out"
L = read_matrix(filename)
```
