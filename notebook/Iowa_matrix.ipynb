```python
def read_matrix(filename):
    

    L3 = []
    for x in range(101):
        L3.append([0]*101)
    file_obj = open(filename,'r')
    for line in file_obj:
        line = line.rstrip()
        fields = line.split()
        #print test2
        #print fields
        x = int(fields[0].split('-')[0])
        y = int(fields[0].split('-')[1])
        if x<101 and y<101:
            L3[x][y] = int(fields[1])
    file_obj.close()
    #print "test"
    #print L1
    #print L2
    #print L3
    #print dimension
    
    return L3


            
            
```

```python
filename = "../data/two_files_matrix.out"
L = read_matrix(filename)
```

how many reads are there that can be co-assembled?
===
criteria:

0 < coverage in sampleA < 5

0 < coverage in sampleB < 5

coverage in sampleA + coverage in sampleB >=5

cutoff = 5, 10, 20


```python
 
def sum_x(cutoff):
    for x in range(1,cutoff):
        sum = 0
        for y in range(1,cutoff):
            if x+y>=cutoff:
                sum = sum+ L[x][y]
    return sum
print "5:",sum_x(5)
print "10:",sum_x(10)
print "20:",sum_x(20)
```
